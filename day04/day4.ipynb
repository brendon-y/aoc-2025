```python
def count(input: list[str]):
    height = len(input)
    width = len(input[0])
    counts: dict[tuple[int, int], int] = dict()
    # print(height, width)
    for i in range(len(input)):
        for j in range(width):
            # print(input[i][j])
            if input[i][j] != '@':
                continue
            neighbours = [(i+ii, j+jj)  
                for ii in range(-1, 2)
                for jj in range(-1, 2)
                    if 0<= i+ii < height and 
                        0<= j+jj < width and
                        not (ii == 0 and jj == 0) and
                        input[i+ii][j+jj] == '@']
            # print(i, j, neighbours)
            counts[(i, j)] = len(neighbours)
                
    # print([(key, value) for key, value in counts.items() if value < 4 and input[key[0]][key[1]] == '@'])
    return len([key for key, value in counts.items() if value < 4 and input[key[0]][key[1]] == '@'])
                        
```

```python
def remove(input: list[str]):
    height = len(input)
    width = len(input[0])
    output: list[str] = []
    # print(height, width)
    for i in range(len(input)):
        line = ''
        for j in range(width):
            # print(input[i][j])
            if input[i][j] != '@':
                line += '.'
                continue
            neighbours = [(i+ii, j+jj)  
                for ii in range(-1, 2)
                for jj in range(-1, 2)
                    if 0<= i+ii < height and 
                        0<= j+jj < width and
                        not (ii == 0 and jj == 0) and
                        input[i+ii][j+jj] == '@']
            if len(neighbours) < 4:
                line += '.'
            else:
                line += '@'
        # print(line)
        output.append(line)
    return output
```

```python
def remove_all(input: list[str]):
    step = remove(input)
    if (step == input):
        return input
    else:
        return remove_all(step)
```

```python
final = remove_all(lines)
flattened = [item for sublist in final for item in sublist]

starting = [item for sublist in lines for item in sublist]
print(starting.count('@') - flattened.count('@'))
```
