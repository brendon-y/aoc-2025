```python
def readFile(name: str):
    with open(name, 'r') as f:
        lines = [line.strip() for line in f]
        return [[int(val) for val in line.split(',')] for line in lines]
        
```

```python
import itertools

def process1(name: str):
    values = readFile(name)
    possibilities = itertools.combinations(values, 2)
    # print(possibilities)
    areas = [(abs(first[0]-second[0]) + 1)*(abs(first[1] - second[1])+1) for [first, second] in possibilities]
    return sorted(areas, reverse=True)
```

```python
import numpy as np

def plotPoints(points: list[list[int]]):
    # print([p[0] for p in points])
    sizeX = max([p[0] for p in points]) + 3
    sizeY = max([p[1] for p in points]) + 2
    board = np.zeros((sizeY, sizeX))
    for p in points:
        board[p[1], p[0]] = 1
    return board

def drawLine(board, pi, pj):
    if (pi[0] == pj[0]):
        ys = sorted([pi[1], pj[1]])
        for y in range(ys[0]+1, ys[1]):
            board[y, pi[0]] = 2
    else:
        xs = sorted([pi[0], pj[0]])
        for x in range(xs[0]+1, xs[1]):
            board[pi[1], x] = 2

def outline(board, points):
    for i in range(len(points) - 1):
        pi = points[i]
        pj = points[i+1]
        drawLine(board, pi, pj)
    drawLine(board, points[-1], points[0])

def fill(board):
    # set background to -1 for now
    yrange = range(len(board))
    xrange =  range(len(board[0]))
    for y in yrange:
        for x in xrange:
            if board[y, x] == 0:
                board[y, x] = -1
    # seed the first pixel as background
    board[0, 0] = 0
    # fill neighbouring pixels
    def areAnyNeighboursBackground(x, y):
        return (x-1 in xrange and board[y, x-1] == 0) or \
            (x+1 in xrange and board[y, x+1] == 0) or \
            (y-1 in yrange and board[y-1, x] == 0) or \
            (y+1 in yrange and board[y+1, x] == 0)

    def enlarge():
        changed = False
        for y in yrange:
            for x in xrange:
                if (board[y, x] == -1 and areAnyNeighboursBackground(x, y)):
                    board[y, x] = 0
                    changed = True
        return changed

    goAgain = enlarge()
    while goAgain:
        goAgain = enlarge()
```
