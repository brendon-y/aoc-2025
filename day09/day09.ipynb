# Part 1

```python
def readFile(name: str):
    with open(name, "r") as f:
        lines = [line.strip() for line in f]
        return [[int(val) for val in line.split(",")] for line in lines]
```

```python
import itertools


def process1(name: str):
    values = readFile(name)
    possibilities = itertools.combinations(values, 2)
    # print(possibilities)
    areas = [
        (abs(first[0] - second[0]) + 1) * (abs(first[1] - second[1]) + 1)
        for [first, second] in possibilities
    ]
    return sorted(areas, reverse=True)
```

```python
print(process1("small.txt")[0])
process1("input.txt")[0]
```

```python
import shapely
import itertools


def readFile(name: str):
    with open(name, "r") as f:
        lines = [line.strip() for line in f]
        return [[int(val) for val in line.split(",")] for line in lines]


points = readFile("input.txt")
mosaic = shapely.Polygon(points)

possibilities = itertools.combinations(points, 2)

maxArea = 0
for p1, p2 in possibilities:
    minx = min(p1[0], p2[0])
    maxx = max(p1[0], p2[0])
    miny = min(p1[1], p2[1])
    maxy = max(p1[1], p2[1])
    rectangle = shapely.Polygon(
        [[minx, miny], [maxx, miny], [maxx, maxy], [minx, maxy]]
    )
    if mosaic.contains(rectangle):
        maxArea = max(maxArea, (maxx - minx + 1) * (maxy - miny + 1))

print(maxArea)
```
