# Part 1

```python
def readFile(name):
    with open(name, 'r') as f:
        lines = [line.strip() for line in f.readlines()]
    return lines
```

```python
def processFile(name):
    lines = readFile(name)
    foundBlank = False
    fresh = []
    good = []
    for line in lines:
        if len(line) == 0:
            foundBlank = True
            continue
        if not foundBlank:
            values = [int(x) for x in line.split('-')]
            fresh.append((values[0], values[1]))
        else:
            value = int(line)
            for collection in fresh:
                if collection[0] <= value <= collection[1]:
                    good.append(value)
                    break
    return good, fresh
```

```python
(good, fresh) = processFile('input.txt')
print(len(good))

(smallGood, smallFresh) = processFile('small.txt')
print(len(smallGood))
```
