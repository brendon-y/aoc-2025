```python
from dataclasses import dataclass
import numpy as np
import numpy.typing as npt
from typing import Any


@dataclass
class slot:
    shape: str
    shapes: list[int]


@dataclass
class shapeType:
    data: np.ndarray
    pixels: int


def read(name):
    with open(name, "r") as f:
        lines = f.readlines()

    shapes: dict[int, shapeType] = {}
    slots: list[slot] = []

    i = 0
    shapeN = []
    pixels = 0
    index = 0

    while "x" not in lines[i]:
        line = lines[i].strip()
        if len(line) == 0:
            shapes[index] = shapeType(np.array(shapeN), pixels)
        if ":" in line:
            # print(line[:-1])
            index = int(line[:-1])
            shapeN = []
            pixels = 0
        else:
            p = [1 if ch == "#" else 0 for ch in line]
            shapeN.append(p)
            pixels += np.add.reduce(p)

        i += 1

    while i < len(lines):
        line = lines[i].strip()
        parts = line.split()
        slots.append(slot(parts[0][:-1], [int(x) for x in parts[1:]]))
        i += 1

    return shapes, slots
```

```python
shapes, slots = read("small.txt")
print(shapes)
print(slots)
```

```python
# for a first approximation, can we just naively lay out the tiles assuming 
# that they can fit without gaps?

for slot in slots:
    area = np.multiply.reduce([int(i) for i in slot.shape.split('x')])
    shapeAreas = [shapes[shapeName].pixels for shapeName in slot.shapes]
    consumed = np.add.reduce(shapeAreas)
    print(area, slot.shapes, shapeAreas, consumed)
```
