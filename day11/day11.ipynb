# Part 1

```python
def getGraph(name: str):
    graph = {}
    with open(name, "r") as f:
        for line in f:
            values = line.strip().split()
            source = values[0][:-1]
            graph[source] = values[1:]
    return graph
```

```python
graph = getGraph("input.txt")
count = 0


def trace(node, target):
    global count
    if node == target:
        count += 1
    elif node in graph:
        children = graph[node]
        for child in children:
            trace(child, target)


trace("svr", "out")
print(count)
```

# Part 2

```python
from dataclasses import dataclass
import functools
import numpy as np


@dataclass(frozen=True)
class blah:
    node: str
    target: str
    hasDac: bool = False
    hasFft: bool = False


graph2 = getGraph("input.txt")


@functools.cache
def trace2(inputs: blah):
    # print(inputs.node)
    if inputs.node == inputs.target:
        if inputs.hasDac and inputs.hasFft:
            return 1
        return 0

    newHasDac = inputs.hasDac or inputs.node == "dac"
    newHasFft = inputs.hasFft or inputs.node == "fft"

    return np.add.reduce(
        [
            trace2(blah(child, inputs.target, newHasDac, newHasFft))
            for child in graph2[inputs.node]
        ]
    )


trace2(blah("svr", "out", False, False))
```
