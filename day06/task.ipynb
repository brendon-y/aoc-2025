# Part 1

```python
import numpy as np
def fetch(name: str):
    with open(name, 'r') as f:
        raw_input = [line.split() for line in f]
    return np.array(raw_input)
```

```python
def process(name: str, parse):
    first = fetch(name)
    sum = 0
    for col in range(len(first[0])):
        entries = first[:, col]
        values = parse(entries[:-1])
        operator = entries[-1]
        result = 0
        match operator:
            case '+':
                result = np.add.reduce(values)
            case '*':
                result = np.multiply.reduce(values)
        # print(values, operator, result)
        sum += result
    return sum
```

```python
print(process('small.txt', lambda entries: [int(x) for x in entries]))
print(process('input.txt', lambda entries: [int(x) for x in entries]))
```

```python
def process(name:str):
    with open(name, 'r') as f:
        lines =  [line for line in f]
    length = max([len(line) for line in lines])
    sum = 0
    operator = '+'
    entries = []

    for x in range(length):
        row = [line[x] if len(line) >  x else ' ' for line in lines]
        last = row[-1]
        if last == '+' or last == '*':
            operator = last
            row = row[:-1]
        row = ''.join(row).strip()
        if row == '':
            # do the thing
            # print('got it', entries, operator)
            if operator == '+':
                sum += np.add.reduce(entries)
            else: 
                sum += np.multiply.reduce(entries)
            entries = []
        else:
            entries.append(int(row))
    return sum
        
        
process('small.txt')
process('input.txt')
    
```
