# Part 1

```python
def readFile(name:str):
    with open(name, 'r') as f:
        lines = [line.strip() for line in f]
    return lines
```

```python
def countEm(name:str):
    lines = readFile(name)
    nextLine = list(lines[0])
    # print(''.join(nextLine))
    splits = 0
    for i in range(len(lines)-1):
        inputLine = nextLine
        inputNextLine = lines[i+1]
        nextLine = ['.' for i in range(len(inputLine))]
        for j in range(len(inputLine)):
            match(inputLine[j], inputNextLine[j]):
                case ('S', '.'):
                    nextLine[j] = '|'
                case ('|', '.'):
                    nextLine[j] = '|'
                case ('|', '^'):
                    nextLine[j-1] = '|'
                    nextLine[j+1] = '|'
                    nextLine[j] = inputNextLine[j]
                    splits += 1
        # print(''.join(nextLine), nextLine.count('|'), splits)
    return(splits)
```

```python
print(countEm('small.txt'))
print(countEm('input.txt'))
```

# Part 2

```python
def parse(ch: str):
    try:
        if (ch == 'S'):
            return 1
        if (ch == '^'):
            return -1
        return int(ch)
    except:
        return 0

def convert(line: str):
    return [parse(ch) for ch in line]

import numpy as np

def countEmQuantum(name:str):
    lines = readFile(name)
    nextLine = convert(lines[0])
    splits = 0
    for i in range(len(lines)-1):
        inputLine = nextLine
        inputNextLine = convert(lines[i+1])
        nextLine = [0 for i in range(len(inputLine))]
        for j in range(len(inputLine)):
            if inputNextLine[j] == -1 and inputLine[j] != 0:
                nextLine[j-1] += inputLine[j]
                nextLine[j+1] += inputLine[j]
                nextLine[j] = 0
            elif inputNextLine[j] == -1:
                nextLine[j] = 0
            else:
                nextLine[j] += inputLine[j]
        # print(nextLine)
          
    return np.add.reduce([val for val in nextLine if val > 0])
```

```python
print(countEmQuantum('small.txt'))
print(countEmQuantum('input.txt'))
```
